```python
def back(n,one_str,i,k,list):
    if( i == n):
        sum = 0
        for i in one_str:
            sum += i
        if (sum == k):
            list.append(one_str)
        return

    one_str[i] = 0
    back(n,one_str,i+1,k,list)

    one_str[i] = 1
    back(n,one_str,i+1,k,list)



n = 15
k = 5
back(n,one_str,0,k,list)
        
```
